```python
# Converting binary variable to 0 and 1
car_deatils['enginelocation'] = car_deatils['enginelocation'].map({'front': 1, 'rear':0})
car_deatils['fueltype'] = car_deatils['fueltype'].map({'gas': 1, 'diesel':0})
car_deatils['aspiration'] = car_deatils['aspiration'].map({'std': 1, 'turbo':0})
car_deatils['doornumber'] = car_deatils['doornumber'].map({'four': 1, 'two':0})
car_deatils.head()
```

```python
#Removing Car company name as it is not making any significance
car_deatils = car_deatils.drop(['car_ID'], axis = 1) 
```

Creating dummy variabls - 
    - carbody
    - drivewheel
    - enginetype (many)
    - cylindernumber
    - fuelsystem

```python
# Creating dummy variables using pandas for some of the feature which are relavent to create dummy
carbody = pd.get_dummies(car_deatils['carbody'], drop_first = True)
car_deatils = pd.concat([car_deatils, carbody], axis = 1)
car_deatils.drop(['carbody'], axis = 1, inplace = True)

drivewheel = pd.get_dummies(car_deatils['drivewheel'], drop_first = True)
car_deatils = pd.concat([car_deatils, drivewheel], axis = 1)
car_deatils.drop(['drivewheel'], axis = 1, inplace = True)

enginetype = pd.get_dummies(car_deatils['enginetype'], drop_first = True)
car_deatils = pd.concat([car_deatils, enginetype], axis = 1)
car_deatils.drop(['enginetype'], axis = 1, inplace = True)

cylindernumber = pd.get_dummies(car_deatils['cylindernumber'], drop_first = True)
car_deatils = pd.concat([car_deatils, cylindernumber], axis = 1)
car_deatils.drop(['cylindernumber'], axis = 1, inplace = True)

fuelsystem = pd.get_dummies(car_deatils['fuelsystem'], drop_first = True)
car_deatils = pd.concat([car_deatils, fuelsystem], axis = 1)
car_deatils.drop(['fuelsystem'], axis = 1, inplace = True)

CarName = pd.get_dummies(car_deatils['CarName'], drop_first = True)
car_deatils = pd.concat([car_deatils, CarName], axis = 1)
car_deatils.drop(['CarName'], axis = 1, inplace = True)

car_deatils.head()
```

```python
# Importing RFE and LinearRegression lib
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Creating a list of columns which we will be using for our model creation that we got it from RFE
col = X_train.columns[rfe.support_]
col
```

```python
# columns which we are not considering for our model creation
X_train.columns[~rfe.support_]
```

# Buliding model using states model

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

# Iteration 1

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the linear model
lm = sm.OLS(y_train,X_train_rfe).fit()   
```

```python
#Printing the summary of our linear model to check for the r-squared and p value of selected variables.
print(lm.summary())
```

From the above model summary we see that r-squared is very good but having features for them p-value is greater than 0.05, so we need to remove these variables as we cant keep in our model 

```python
# Calculate the VIFs for the new model, 
# VIF is also part of selecting variables what all features we need to keep to get the best model
# A feature having greater than 5 vif, can not be part of our model as it shows that it is correlated with some other 
# variable and redundent, so we need to remove these variables
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Below action should be performd, If created model has -
    - High p-value and High VIF -> needs to be removed first
    - High p-value and Low VIF -> needs to be removed first column having high p-value
    - Low p-value and High VIF -> needs to be removed first other column having high p-value, then check for the VIF
    - Low p-value and Low VIF -> should not be removed 
    

```python
# Dropping the high p value first
# subaru is insignifisent for our model as having greater p value hence droping it.
X_train_new = X_train_rfe.drop(["subaru"], axis = 1)
```
